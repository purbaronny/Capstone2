## **3. Explanatory Data Analyst (EDA)**

```python
def set_labels(x="", y="", title="", axis=None):
  if axis:
    axis.set_xlabel(x);
    axis.set_ylabel(y);
    axis.set_title(title);
  else:
    plt.title(title);
    plt.xlabel(x);
    plt.ylabel(y);
```

### **3.1. People**

#### **3.1.1 Bagaimana distribusi tahun lahir pelanggan kami?**

```python
# Data
data = df['Year_Birth']
bins = list(range(1830, 2005, 5))  # Sesuaikan dengan kebutuhan visual

# Hitung frekuensi
counts, bin_edges = np.histogram(data, bins=bins)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

# Plot
plt.figure(figsize=(16, 6)) 
sns.histplot(data, bins=bins, edgecolor='black')  # Gunakan bins yang sama

# Tambahkan teks jumlah
for count, x in zip(counts, bin_centers):
    if count > 0:
        plt.text(x, count + 1, str(count), ha='center', va='bottom', fontsize=8)

# Label dan judul
set_labels(x="Tahun Lahir", y="Frequency", title="Distribusi Usia Konsumen")
plt.xticks(list(range(1830, 2000, 5)))
plt.tight_layout()
plt.show()
```

**Berdasarkan data:**
1. Sebagian besar pelanggan berusia antara tahun 1970 hingga 1978.
2. Orang tua tampaknya memiliki frekuensi lebih banyak daripada individu yang lebih muda.
3. Ada beberapa pengecualian.

#### **3.1.2 Bagaimana distribusi tingkat pendidikan?**

```python
# Urutan kategori sesuai permintaan
edu_order = ["Basic", "2nd Cycle", "Graduation", "Master", "PhD"]

plt.figure(figsize=(12, 6))
ax = sns.countplot(x=df.Education, order=edu_order)

# Tambahkan jumlah di atas tiap batang
for p in ax.patches:
    count = int(p.get_height())
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.text(x, y + 1, str(count), ha='center', va='bottom', fontsize=10)

# Set label dan judul
set_labels(x="Tingkat Pendidikan", y="Frequency", title="Tingkat Pendidikan Diantara Pelanggan", axis=ax)

plt.tight_layout()
plt.show()
```

**Berdasarkan data:**

1. Sebagian besar pelanggan memiliki tingkat pendidikan Graduation (Sarjana).
2. Tingkat pendidikan Basic dan 2nd Cycle masih cukup umum, menunjukkan sebagian pelanggan belum menempuh pendidikan tinggi.
3. Pelanggan dengan gelar Master dan PhD memiliki frekuensi lebih sedikit dibandingkan tingkat pendidikan lainnya.
4. Hal ini menunjukkan bahwa mayoritas pelanggan telah menyelesaikan pendidikan menengah atau tinggi, namun hanya sebagian kecil yang melanjutkan ke pascasarjana.
5. Ada kemungkinan bahwa pelanggan dengan pendidikan lebih tinggi memiliki karakteristik atau kebutuhan yang berbeda dalam perilaku konsumennya.

#### **3.1.3 What is the martial status of the majority of customers?**

```python
# Ambil data
rel = df.Marital_Status.value_counts()

# Plot diagram batang
plt.figure(figsize=(8, 5))
ax = sns.barplot(x=rel.index, y=rel.values, hue=rel.index, palette='pastel', edgecolor='black', legend=False)

# Tambahkan jumlah di atas batang
for i, v in enumerate(rel.values):
    plt.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=10)

# Set label dan judul
set_labels(x="Status Perkawinan", y="Frekuensi", title="Distribusi Status Perkawinan di Antara Pelanggan", axis=ax)

plt.tight_layout()
plt.show()
```

Kita dapat melihat banyak nilai yang tidak valid. **['Menikah', 'Bersama']** dapat dihitung sebagai **Kawin**. 
Sementara **['Lajang', 'Bercerai', 'Janda', 'Sendiri']** semuanya termasuk dalam status **Lajang**. 
Sementara itu 'Tidak masuk akal' dan 'YOLO' dapat dibuang karena tidak memiliki makna apa pun

```python
single = rel[['Single', 'Divorced', 'Widow', 'Alone']].sum()
relationship =  rel[['Married', 'Together']].sum()

set_labels(title="Status Perkawinan Pelanggan")
plt.pie([single, relationship], labels = ['Lajang', 'Kawin'], autopct='%.0f%%');
```

**Berdasarkan Data:**
1. Status Perkawinan Konsumen sebagian besar pelanggan adalah Kawin
